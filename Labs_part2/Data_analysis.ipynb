```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define file path
file_path = "25-03-04-16-20.data"  # Replace with actual file path

# Initialize lists for x and y
x = []
y = []

# Read the file and extract columns
with open(file_path, "r", encoding="utf-8", errors="ignore") as file:
    for line in file:
        parts = line.strip().split()  # Split by whitespace
        if len(parts) == 2:  # Ensure there are exactly two columns
            x.append(parts[0])  # Store first column in x
            y.append(parts[1])  # Store second column in y

# Convert lists to numpy arrays and then to integers
x = np.array(x, dtype=int)
y = np.array(y, dtype=int)

# Print preview
print("x:", x[:5])
print("y:", y[:5])



```

```python
# Filter out values in x that are >= 40000
mask = x < 40000
cleaned_x = x[mask]
cleaned_y = y[mask]
```

```python
# Print preview
print("cleaned_x:", cleaned_x)
print("cleaned_y:", cleaned_y)
print("Length of cleaned_x:", len(cleaned_x))
print("Length of cleaned_y:", len(cleaned_y))
```

```python
# Plot the data
plt.figure(figsize=(12, 6))
plt.hist(cleaned_x, bins=50, color="skyblue", edgecolor="black")
```

```python
offset_x = cleaned_x + 30.977


thresh_t = 3330
# Filter out values in x that are >= 3330
mask = offset_x > thresh_t
thresh_x = offset_x[mask]
thresh_y = cleaned_y[mask]
print("Length of thresh_x:", len(thresh_x))

```

```python
#plot the data
plt.figure(figsize=(12, 6))
plt.hist(thresh_x, bins=40, color="skyblue", edgecolor="black")
plt.title("Histogram of decay values")
#fit an exponential to the data

# Add labels
plt.xlabel("Decay time (ns)")
plt.ylabel("Frequency")
```

```python
time_diff = np.zeros(len(cleaned_y))

for i in range(len(cleaned_y)-1):
    time_diff[i] = cleaned_y[i+1] - cleaned_y[i]


print("cleaned_time:", time_diff)
#plot the data
plt.figure(figsize=(12, 6))
plt.hist(time_diff, bins=40, color="skyblue", edgecolor="black")
plt.title("Histogram of time difference values")

```
